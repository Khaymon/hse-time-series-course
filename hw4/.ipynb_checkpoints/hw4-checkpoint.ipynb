```python
import numpy as np
import matplotlib.pyplot as plt
```

Сгенерируем исходный ряд.

```python
fs = 500
t_end = 2
t = np.linspace(0, t_end, int(fs * t_end), endpoint=False)

freq1 = 5
freq2 = 50
freq3 = 120

signal = 2 * np.sin(2 * np.pi * freq1 * t) + \
         1.0 * np.sin(2 * np.pi * freq2 * t) + \
         1.5 * np.sin(2 * np.pi * freq3 * t)

plt.figure(figsize=(14, 4))
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()
```

И добавим в него нормальный шум.

```python
noise = np.random.normal(0, 2, signal.shape)
noisy_signal = signal + noise

plt.figure(figsize=(14, 4))
plt.plot(t, noisy_signal)
plt.title('Noisy Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()
```

```python
freq_domain = np.fft.fft(noisy_signal)
freqs = np.fft.fftfreq(len(noisy_signal), d=1/fs)
```

Занулим значения в frequency-domain, которые по абсолютам меньше 200.

```python
def denoise_signal(noisy_signal, cutoff_abs):
    freq_domain = np.fft.fft(noisy_signal)
    mask = np.abs(freq_domain) > cutoff_abs
    filtered_freq_domain = freq_domain * mask
    denoised_signal = np.fft.ifft(filtered_freq_domain).real
    estimated_noise = noisy_signal - denoised_signal

    return denoised_signal, estimated_noise
```

```python
denoised_signal, estimated_noise = denoise_signal(noisy_signal, 200)
```

```python
plt.figure(figsize=(14, 10))

plt.subplot(3, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(t, noisy_signal)
plt.title('Noisy Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(t, denoised_signal)
plt.title('Denoised Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

Разность между шумным и почищенным рядом, т.е. оценка на шум.

```python
plt.figure(figsize=(14, 4))
plt.plot(t, estimated_noise)
plt.title('Estimated Noise Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()
```

```python
freqs = np.fft.fftfreq(len(noisy_signal), d=1/fs)
fft_noisy = np.fft.fft(noisy_signal)
fft_denoised = np.fft.fft(denoised_signal)

mask = freqs >= 0
freqs = freqs[mask]
fft_noisy = np.abs(fft_noisy)[mask]
fft_denoised = np.abs(fft_denoised)[mask]

plt.figure(figsize=(14, 6))
plt.subplot(2,1,1)
plt.plot(freqs, fft_noisy)
plt.title('Frequency Spectrum of Noisy Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')

plt.subplot(2,1,2)
plt.plot(freqs, fft_denoised)
plt.title('Frequency Spectrum of Denoised Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")
```

Возьмем датасет количества пассажиров одной из авиакомпаний за определенный промежуток времени.

```python
data = pd.read_csv("AirPassengers.csv")

data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

data.columns = ['AirPassengers']

data.head()
```

```python
# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['AirPassengers'])
plt.title('Air Passengers Time Series')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Decompose the time series
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(data['AirPassengers'], model='multiplicative', period=12)

# Plot the decomposition
fig = decomposition.plot()
fig.set_size_inches(14, 10)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

print("p-value:", adfuller(data['AirPassengers'])[1])
```

На уровне значимости 0.05 гипотеза о нестационарности исходного ряда не отвергается.

Прологарифмируем исходный ряд, чтобы линеаризовать его дисперсию.

```python
data['LogPassengers'] = np.log(data['AirPassengers'])

# Plot the log-transformed data
plt.figure(figsize=(12,6))
plt.plot(data.index, data['LogPassengers'])
plt.title('Log-Transformed Air Passengers Time Series')
plt.xlabel('Year')
plt.ylabel('Log(Number of Passengers)')
plt.show()
```

Найдем разности первого и второго порядков, чтобы сделать ряд стационарным.

```python
data['LogPassengers_diff'] = data['LogPassengers'] - data['LogPassengers'].shift(1)
data_diff = data['LogPassengers_diff'].dropna()

data['LogPassengers_2nd_diff'] = data_diff - data_diff.shift(1)
data_2nd_diff = data['LogPassengers_2nd_diff'].dropna()                                       

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.plot(data_diff)
plt.title('First Differenced Log-Transformed Data')
plt.xlabel('Year')
plt.ylabel('Difference of Log(Number of Passengers)')
plt.subplot(122)
plt.plot(data_2nd_diff)
plt.title('Second Differenced Log-Transformed Data')
plt.xlabel('Year')
plt.ylabel('2nd Difference of Log(Number of Passengers)')
plt.tight_layout()
plt.show()
```

```python
print("p-value:", adfuller(data_diff)[1])
print("p-value:", adfuller(data_2nd_diff)[1])
```

На уровне значимости 0.05 гипотеза о нестационарности разностного ряда второго порядка отвергается. Поэтому, параметром `d = 2` для ARIMA.

Чтобы определиться с параметрами `q` и `p` посмотрим на графики ACF и PACF.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
plt.figure(figsize=(14,6))
plt.subplot(121)
plot_acf(data_diff, lags=40, ax=plt.gca())
plt.subplot(122)
plot_pacf(data_diff, lags=40, ax=plt.gca())
plt.tight_layout()
plt.show()
```

Как видно, наибольшие по абсолютам значения достигаются в точке `12`, поэтому возьмем `q = p = 12`.

```python
import statsmodels
```

```python
model = statsmodels.tsa.arima.model.ARIMA(data['LogPassengers'],
                                  order=(12, 2, 12),
                                  enforce_stationarity=False,
                                  enforce_invertibility=False)
results = model.fit()

results.plot_diagnostics(figsize=(15, 12))
plt.show()
```

По графикам видно, что модель хорошо выучила зависимость в ряде, о чем говорит Q-Q график ошибок. Также видно, что ошибки не автоскоррелированы.

Построим предсказания на следующие 2 года.

```python
pred_uc = results.get_forecast(steps=24)
pred_ci = pred_uc.conf_int()

predicted_mean = np.exp(pred_uc.predicted_mean)
pred_ci_exp = np.exp(pred_ci)

plt.figure(figsize=(14, 7))
plt.plot(data.index, data['AirPassengers'], label='Observed')
plt.plot(predicted_mean.index, predicted_mean, label='Forecast')
plt.fill_between(pred_ci_exp.index, pred_ci_exp.iloc[:, 0], pred_ci_exp.iloc[:, 1], color='k', alpha=.25)
plt.legend()
plt.title('Forecasts for Air Passengers')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```
